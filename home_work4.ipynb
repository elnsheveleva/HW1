```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

# Пункт 1. Пользователь по команде “p” может узнать владельца документа по его номеру
def name_by_number():
    doc = input('Введите номер документа: ')
    for rec in documents:         
        if rec['number'] == doc:
            print('Владелец документа: ' + rec['name'])
            return
    print('Документ не найден')
            
# Пункт 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится
def directories_number():
    doc = input('Введите номер документа: ')
    for directory in directories:
        list = directories[directory]
        for doc_number in list:
            if doc_number == doc:
                print('Документ хранится на полке: ' + directory)
                return
    print('Документ не найден')

# Пункт 3. Пользователь по команде “l” может увидеть полную информацию по всем документам
def all_info():
    for rec in documents:
        rec_type = rec['type'] 
        rec_num = rec['number']
        rec_name = rec['name']
        
        for directory in directories:
            list = directories[directory]
            for doc_number in list:
                if doc_number == rec_num:
                    print('№: ' + rec_num + ', тип: ' + rec_type + ', владелец: ' + rec_name + ', полка хранения: ' + directory)

# Пункт 4. Пользователь по команде “as” может добавить новую полку
def print_directories():
    all_dirs = []
    for directory in directories:
        all_dirs.append(directory)
    return ', '.join(all_dirs)

def add_directory():
    direct_num = input('Введите номер полки: ')
    
    for directory in directories:
        if directory == direct_num:
            print('Такая полка уже существует. Текущий перечень полок: ' + print_directories())
            return
    
    directories[direct_num] = []
    print('Полка добавлена. Текущий перечень полок: ' + print_directories()) 
    
# Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)
def del_directory():
    direct_num = input('Введите номер полки: ')
    
    for directory in directories:
        if directory == direct_num:
            if len(directories[directory]) == 0:
                del directories[directory]
                print('Полка удалена. Текущий перечень полок: ' + print_directories())
            else:
                print('На полке есть документы, удалите их перед удалением полки.'+
                      ' Текущий перечень полок: ' + print_directories())
            return
    print('Такой полки не существует. Текущий перечень полок: ' + print_directories())
    
command = input('Введите команду: ')
while command != 'q':
    if command == 'p':
        name_by_number()
    elif command == 's':
        directories_number()
    elif command == 'l':
        all_info()
    elif command == 'as':
        add_directory()
    elif command == 'ds':
        del_directory() 
    else:
        print('Неизвестная команда')
    command = input('Введите команду: ')
print('Конец')
```
